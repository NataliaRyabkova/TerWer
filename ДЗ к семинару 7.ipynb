```python
# Задача 1.Даны две независимые выборки. Не соблюдается условие нормальности
# x1 380,420, 290
# y1 140,360,200,900
# Сделайте вывод по результатам, полученным с помощью функции
# Для рассчета используем критерий Манна – Уитни.

import scipy.stats as stats
x1=np.array([380,420, 290])
y1=np.array([140,360,200,900])
print(stats.mannwhitneyu(x1,y1))
# Так как pvalue>0,05, то 
print ("На уровне значимости alfa=0.05 статистически значимых различий нет")
```

```python
# Задача 2.Исследовалось влияние препарата на уровень давления пациентов. Сначала измерялось давление до приема препарата, потом через 10 минут и через 30 минут.
# Есть ли статистически значимые различия?
# 1е измерение до приема препарата: 150, 160, 165, 145, 155
# 2е измерение через 10 минут: 140, 155, 150, 130, 135
# 3е измерение через 30 минут: 130, 130, 120, 130, 125
# Для рассчета используем критерий Фридмана.

import scipy.stats as stats
import numpy as np
do_priema = np.array([150, 160, 165, 145, 155])
cherez_10min = np.array([140, 155, 150, 130, 135])
cherez_30min = np.array([130, 130, 120, 130, 125])
print (stats.friedmanchisquare(do_priema,cherez_10min,cherez_30min))
# Так как pvalue<0,05, то 
print ("На уровне значимости alfa=0.05 статистически значимые различия есть")
```

```python
# Задача 3. Сравните 1 и 2 е измерения, предполагая, что 3го измерения через 30 минут не было.
# Исследовалось влияние препарата на уровень давления пациентов. Сначала измерялось давление до приема препарата, потом через 10 минут
# 1е измерение до приема препарата: 150, 160, 165, 145, 155
# 2е измерение через 10 минут: 140, 155, 150, 130, 135
# Для рассчета используем критерий Уилкоксона.

do_priema = np.array([150, 160, 165, 145, 155])
cherez_10min = np.array([140, 155, 150, 130, 135])
array = cherez_10min-do_priema
print("Разница между значениями после приема препарата через 10 минут и до приема",array)
print (stats.wilcoxon(do_priema,cherez_10min))
# Так как pvalue>0,05, то 
print ("На уровне значимости alfa=0.05 статистически значимых различий нет")
```

```python
# Задача 4. Даны 3 группы учеников плавания.
# В 1 группе время на дистанцию 50 м составляют: 56, 60, 62, 55, 71, 67, 59, 58, 64, 67
# Вторая группа : 57, 58, 69, 48, 72, 70, 68, 71, 50, 53
# Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54
# Для рассчета используем критерий Крускала-Уоллиса.

gr_1=np.array([56, 60, 62, 55, 71, 67, 59, 58, 64, 67])
gr_2=np.array([57, 58, 69, 48, 72, 70, 68, 71, 50, 53])
gr_3=np.array([57, 67, 49, 48, 47, 55, 66, 51, 54])
print (stats.kruskal(gr_1,gr_2,gr_3))
# Так как pvalue>0,05, то 
print ("На уровне значимости alfa=0.05 статистически значимых различий нет")
```
