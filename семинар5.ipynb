```python
# Задача 2. Проведите тест гипотезы. Утверждается, что шарики для подшипников, изготовленные автоматическим станком, имеют средний 
# диаметр 17 мм. Используя односторонний критерий с α=0,05, проверить эту гипотезу, если в выборке из n=100 шариков средний диаметр 
# оказался равным 17.5 мм, а дисперсия известна и равна 4 кв. мм.
# h0: средний диаметр шариков = 17 мм
# h1: средний диаметр шариков > 17 мм (одностронний критерий)
import scipy.stats as stats
a = 0.05
n = 100
mu0 = 17
mu1 = 17.5 
D = 4
std = D ** 0.5
z1 = (mu1 - mu0) / (std / (n ** 0.5))
print ('Наблюдаемое значение z-критерия:',z1)
# Из таблицы получаем,что при  0,95 значение z0 = 1.645 
z0 = 1.645
# Проверка:
zt = stats.norm.ppf(1 - 0.05)
print ('Табличное значение z-критерия:',z0, 'Проверка:', zt)
print ('Для a = 0.05 нулевую гипотезу отклоняем, так как  z1>z0')
```

```python
# Задача 3. Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.
# Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет 202, 203, 199, 197, 195, 201, 200, 204, 194, 190. Известно, что их веса распределены нормально.
# Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? (Провести двусторонний тест.)

# h0: средний вес = 200
# h1: средний вес != 200 - не равен (двухсторонний критерий)
cookie = np.array([202, 203, 199, 197, 195, 201, 200, 204, 194, 190])
p=0.99
# Будем использовать двусторонний критерий c уровнем значимости a=1-p=1-0.99=0.01
a=0.01
al = a/2 
n=10
m0 = 200
# Из таблицы находим для а = 0,01 и (n-1) = 9, получаем t0 = 3.25 
t0=3.25
print ('Табличное значение t-критерия:', t0, 'Проверка:', stats.t.ppf((1-al),len(cookie)-1))
# Найдем мат.ожидание:
m1 = cookie.mean()
# Найдем среднеквадратичное отклонение
std = cookie.std(ddof=1)
t1 = (m1 - m0) / (std / (n ** 0.5))
print ('Наблюдаемое значение t-критерия',t1,'Проверка:', (np.mean(cookie)-200)/(np.std(cookie,ddof=1)/len(cookie)**0.5))
# Сравним полученные значения (так как двусторонний критерий, то используем абсолютное значение полученного t-критерия)
print ('Для a = 0.01 принимаем нулевую гипотезу, так как |t1| < t0')
```

```python
# Задача 4. Задачу 4 решать с помощью функции. Есть ли статистически значимые различия в росте дочерей?
# 4) Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169
# Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160
print ('В представленной задаче у нас двухвыборочный тест,с зависимыми выборками, решим с помощью  применения функций:') 
# Нулевая гипотеза различий нет - μ=μ0
# Альтернативная гипотеза, есть статистически значимые различия - μ≠ μ0
mum = ([172, 177, 158, 170, 178,175, 164, 160, 169])
daughter =([173, 175, 162, 174, 175, 168, 155, 170, 160])
a=0.05
print (stats.ttest_rel(a = mum,b = daughter))
# Статистика t-теста равна 0.5595, двустороннее значение p равно 0.5911.
print ('Так как p > a, то мы принимаем нулевую гипотезу, то есть статистически значимых различий нет.')
```
