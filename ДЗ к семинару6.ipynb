```python
# Задача 1. Известно, что генеральная совокупность распределена нормально со средним квадратическим отклонением, равным 16.
# Найти доверительный интервал для оценки математического ожидания a с надежностью 0.95,если выборочная средняя M = 80, а объем выборки n = 256.
import math 
import numpy as np
import scipy.stats as stats
s = 16
p = 0.95
M = 80
n = 256

# a=1-p, 1 - a/2 = 0.975
# Находим табличное значение z1, т.е. табличное значение z(a/2) критерия для 95% доверительного интервала:
a=1-p
a1 = 1.96
z = a1 * s / (n ** 0.5)
z1 = stats.norm.ppf(1 - a/2)
print ('Табличное значение z-критерия:',z, 'Проверка:', z1)
x1=M-z
x2=M+z
print(f'Доверительный интервал равен: [{x1};{x2}].')
```

```python
# Задача 2. В результате 10 независимых измерений некоторой величины X, выполненных с одинаковой точностью,получены опытные данные:
# 6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1 Предполагая, что результаты измерений подчинены нормальному закону распределения вероятностей,
# оценить истинное значение величины X при помощи доверительного интервала, покрывающего это значение с доверительной вероятностью 0,95.
from scipy.stats import norm
x = np.array([6.9, 6.1, 6.2, 6.8, 7.5, 6.3, 6.4, 6.9, 6.7, 6.1])
# Размер выборки
n = len(x)
# Среднее квадратическое отклонение по выборке
std = x.std(ddof=1)
# Среднее выборочное
M = x.mean()
print (f'Среднее выборочное: {M}' ,f'Размер выборки: {n}',f'Среднее квадратическое отклонение: {std}.')
# a/2 = 0.05/2 = 0.025, 
# по таблице:
t = 2.262
a = 0.05
t1 = stats.t.ppf(1 - a / 2, df=n - 1)
print ('Табличное значение t-критерия:',t, 'Проверка:', t1)
# Рассчитаем доверительный интервал:
x1=M - t * std / np.sqrt(n)
x2=M + t * std / np.sqrt(n)
print(f'Доверительный интервал равен: [{x1};{x2}].')
```

```python
# Задача 3. Рост дочерей 175, 167, 154, 174, 178, 148, 160, 167, 169, 170
# Рост матерей 178, 165, 165, 173, 168, 155, 160, 164, 178, 175
# Используя эти данные построить 95% доверительный интервал для разности среднего роста родителей и детей.
import scipy.stats as stats
a=np.array([175, 167, 154, 174, 178, 148, 160, 167, 169, 170])
b=np.array([8, 165, 165, 173, 168, 155, 160, 164, 178, 175])
# Размер выборки
na = len(a)
nb = len(b)
# Среднее выборки a
Ma = a.mean()
# Среднее выборки b
Mb = b.mean()
# Разность средних выборок
Delta = Ma-Mb
print (f'Среднее выборки Ma={Ma} и Mb={Mb} ' ,f'Размер выборки a={na} и b={nb}',f'Разность средних выборок равна: {Delta}.')
# Несмещенная дисперсия для a
Da = np.var(a,ddof=1)
# Несмещенная дисперсия для b
Db = np.var(b,ddof=1)
# Объединенная оценка дисперсии
D = (Da+Db)/2
# Стандартная ошибка разности средних
S = np.sqrt(D/na+D/nb)
x=2*(n-1)
t=stats.t.ppf(0.975,x)
# Рассчитаем доверительный интервал:
x1=Delta - S*t
x2=Delta + S*t
print(f'Доверительный интервал равен: [{x1};{x2}].')
```
