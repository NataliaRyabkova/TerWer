```python
# Задача.Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. 
# Даны значения роста в трех группах случайно выбранных спортсменов: Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. 
# Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. 
# Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.
import numpy as np
import pandas as pd
from scipy import stats

f=np.array([173, 175, 180, 178, 177, 185, 183, 182])
h=np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
sh=np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
k=3
n1 = f.shape[0]
n2 = h.shape[0]
n3 = sh.shape[0]
n = n1 + n2 + n3
# Надем средние выборочные по каждой группе
f_m = np.mean(f)
h_m = np.mean(h)
sh_m = np.mean(sh)
print ('Средние выборочные по футболистам,хоккеистам и штангистам соответственно равны',f_m,  h_m,  sh_m )
# Заметим, что средний рост немного отличается. Проверим статистическую значимость этого отличия.
# Объединим все значения в один массив.
nn = np.concatenate([f, h, sh])
n_m = nn.mean()
print('Среднее выборочное по все значениям равно', n_m)
# Найдем сумму квадратов отклонений средних групповых значений от общего среднего:
s2_f = n1 * (f_m -n_m) ** 2 + n2 * (h_m -n_m) ** 2 + n3 * (sh_m - n_m) ** 2
print ('Cумма квадратов отклонений средних групповых значений от общего среднего равна',s2_f )
# Найдем остаточную сумму квадратов отклонений:
s2_r = np.sum((f - f_m)**2) + np.sum((h - h_m)**2) + np.sum((sh - sh_m)**2)
print('Cумма квадратов отклонений',s2_r)
# Найдем сумму квадратов отклонений наблюдений от общего среднего. Проверим выполнение равнства S = s2_f+s2_r
S1=s2_f + s2_r
S2=((nn - n_m) ** 2).sum()
print ('Получаем сумму квадратов отклонений от общего', S1, S2)
# Найдем  общую дисперсию
sigma = S1 / (n - 1)
print('Дисперсия общая равна',sigma)
# Найдем факторную дисперсию
sigma2 = s2_f / (k - 1)
print('Дисперсия факторная равна',sigma2)
# Найдем остаточную дисперсию:
sigma3 = s2_r / (n - k)
print('Дисперсия остаточная равна',sigma3)
# Рассчитаем статистику
T = sigma2 / sigma3
# Проверка
T1=stats.f_oneway(f, h, sh)
print('Значение статистики равно:',T,'Проверка',T1)
# Возьмем уровень значимости равным 0,05. Для него найдём критическое значение
alpha = 0.05
k1 = k - 1
k2 = n - k
F = stats.f.ppf(1 - alpha, k1, k2)
print('Критическое значение равно', F)
# Так как pvalue = 0.01048<alpha = 0.05 отвергаем нулевую гипотезу и получаем,
# что различия среднего роста среди взрослых футболистов, хоккеистов и штангистов есть. И так как T>F, делаем вывод
print('Различия среднего роста среди взрослых футболистов, хоккеистов и штангистов есть, оно является статистически значимым')
```
