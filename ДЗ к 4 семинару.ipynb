```python
# Задача 4. Рост взрослого населения города X имеет нормальное распределение. Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см. 
# Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост: а). больше 182 см б). больше 190 см в). от 166 см до 190 см 
# г). от 166 см до 182 см д). от 158 см до 190 см е). не выше 150 см или не ниже 190 см ё). не выше 150 см или не ниже 198 см ж). ниже 166 см.
from scipy.stats import norm
M = 174
std = 8
def znachenie_z(n):
    return (n-174)/8
# а). больше 182 см
print('больше 182 см, z =', znachenie_z(182))
# По таблице находим 
P = 0.84134
print ('Вероятность равна:', 1-P)
# Проверка: 
pa=1-norm.cdf(182, M, std)
print('Проверка:', pa)
print ()

# б). больше 190 см
print('больше 190 см, z =', znachenie_z(190))
# По таблице находим 
P = 0.9772
print ('Вероятность равна:', 1-P)
# Проверка: 
pb=1-norm.cdf(190, M, std)
print('Проверка:', pb)
print() 

# в). от 166 см до 190 см
print ('Вероятность от 166 см до 190 см')
print('166 см, z1 =', znachenie_z(166))
print('190 см, z2 =', znachenie_z(190))
# По таблице находим 
P1 = 0.1587
P2 = 0.9772
print ('Вероятность равна:', P2-P1)
# Проверка: 
print('Проверка:',norm.cdf(190, M, std)-norm.cdf(166, M, std))
print()

# г). от 166 см до 182 см
print ('Вероятность от 166 см до 182 см')
print('166 см, z1 =', znachenie_z(166))
print('182 см, z2 =', znachenie_z(182))
# По таблице находим 
P1 = 0.1587
P2 = 0.84134
print ('Вероятность равна:', P2-P1)
# Проверка: 
print('Проверка:',norm.cdf(182, M, std)-norm.cdf(166, M, std))
print()

# д). от 158 см до 190 
print ('Вероятность от 158 см до 190 см')
print('158 см, z1 =', znachenie_z(158))
print('190 см, z2 =', znachenie_z(190))
# По таблице находим 
P1 = 0.0228
P2 = 0.9772
print ('Вероятность равна:', P2-P1)
# Проверка: 
print('Проверка:',norm.cdf(190, M, std)-norm.cdf(158, M, std))
print()

# е). не выше 150 см или не ниже 190 см
print ('не выше 150 см или не ниже 190 см')
print('150 см, z1 =', znachenie_z(150))
print('190 см, z2 =', znachenie_z(190))
# По таблице находим 
P1 = 0.0013
P2 = 0.9772
P22 = 1-P2
print ('Вероятность равна:', P1+P22)
# Проверка:
print('Проверка:',((1-norm.cdf(190, M, std)) + norm.cdf(150, M, std)))
print()

# ё). не выше 150 см или не ниже 198 см
print ('не выше 150 см или не ниже 198 см')
print('150 см, z1 =', znachenie_z(150))
print('190 см, z2 =', znachenie_z(198))
# По таблице находим 
P1 = 0.0013
P2 = 0.9986
P22 = 1-P2
print ('Вероятность равна:', P1+P22)
# Проверка:
print('Проверка:',((1-norm.cdf(198, M, std)) + norm.cdf(150, M, std)))
print()

# ж). ниже 166 см.
print ('ниже 166 см')
print('166 см, z =', znachenie_z(166))
# По таблице находим 
Pj = 0.1587
print ('Вероятность равна:', Pj)
# Проверка:
print('Проверка:',norm.cdf(166, M, std))

```
